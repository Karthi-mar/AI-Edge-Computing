```python
import numpy as np

def softmax(x):  #define the softmax activation function of the neuron
    return np.exp(x)/np.sum(np.exp(x))

x=np.array([0.1,0.9,4.0])
output=softmax(x)
print('Softmax in Python :',output)

y_pred=np.array([0.0190013,0.04228816,0.93871054]) #define the predicted input by softmax
y_true=np.array([0.0,0.0,1.0]) #define the actual input

def CE(y_pred,y_true): #define cross entropy loss function
    total_CE_loss=np.sum(-y_true*np.log(y_pred)-(1-y_true)*np.log(1-y_pred))
    num_of_samples=y_pred.shape[0]
    mean_CE_loss=total_CE_loss/num_of_samples
    return mean_CE_loss

CE_value = CE(y_pred,y_true)
print('Cross entropy error is :'+str(CE_value))
```
